# Merging Scraped Restaurant Reviews

#### Purpose
This notebook consolidates all individually scraped review files into
a **single unified dataset**.

It is the final step of the RestaurantGuru scraping pipeline.

---

#### What This Notebook Does
- Loads all CSV files generated during the review scraping process
- Concatenates them into one pandas DataFrame
- Performs basic consistency checks
- Exports the merged dataset as a single CSV file

---

#### Role in the Project
This notebook produces the **final dataset** that can be used for:
- Data analysis
- Sentiment analysis
- Machine learning applications
- Visualization and reporting

---

#### Output
- One consolidated CSV file containing all restaurant reviews

---

#### Notes
- Ensure that all review CSV files are stored in the same directory
- The output file can be easily extended with additional cleaning steps


```python
import os
import glob
import pandas as pd
```

```python
folder_path = r"\Outputs_Bewertungen"

# Erstelle eine leere Liste zum Speichern der DataFrames
dfs = []

# Durchsuche den Ordner nach CSV-Dateien
for filename in glob.glob(os.path.join(folder_path, "*.csv")):
    # Lese jede CSV-Datei als DataFrame ein
    df = pd.read_csv(filename)
    df["Dateiname"]=os.path.basename(filename)
    # FÃ¼ge das DataFrame zur Liste hinzu
    dfs.append(df)

# Kombiniere alle DataFrames in ein einzelnes DataFrame
bewertungen_concat = pd.concat(dfs, ignore_index=True)

```
